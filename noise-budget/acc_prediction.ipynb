```python
# Import all needed libraries and sublibraries

import tensorflow as tf

from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping
from keras import metrics 

import pandas as pd

import sklearn
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
# Import input (x) and output (y) data
# X = pd.read_csv('kernel_acc.csv')
# y = pd.read_csv('combination_acc.csv')
data = pd.read_csv('acc2acc_SSIM_0.99_4k.csv')
X = data.iloc[:,:-1]
y = data.iloc[:,-1:]

# Plot data to visualize the correlation of kernel1/kernel2 to combination
plt.figure()
plt.plot(X.nlm, y, 'b.', label='NLM')
plt.plot(X.dmmhc, y, 'r.', label='DMMHC')
plt.plot(X.gbc, y, 'g.', label='GC')
plt.plot(X.unsharp, y, 'k.', label='Unsharp')
plt.legend()
plt.xlabel('kernel accuracy')
plt.ylabel('combination accuracy')

# plt.figure()
# ax = plt.axes(projection='3d')
# ax.scatter(X.nlm, X.unsharp, y, c='b', marker='o')
# plt.show()
```

```python
# Split the data into input (x) training and testing data, and ouput (y) training and testing data, 
# with training data being 80% of the data, and testing data being the remaining 20% of the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Scale data
X_train_scaled = preprocessing.scale(X_train)
X_test_scaled = preprocessing.scale(X_test)
```

```python
# Defines "deep" model and its structure
model = Sequential()
model.add(Dense(12, input_shape=(4,), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(Adam(lr=0.01), 'mean_squared_error')

# Pass several parameters to 'EarlyStopping' function and assigns it to 'earlystopper'
# earlystopper = EarlyStopping(monitor='val_loss', min_delta=0, patience=15, verbose=1, mode='auto')

# Fits model over 2000 iterations with 'earlystopper' callback, and assigns it to history
# history = model.fit(X_train_scaled, y_train, epochs = 2000, validation_split = 0.2, shuffle = True, verbose = 0, 
#                     callbacks = [earlystopper])
model.fit(X_train_scaled, y_train, epochs=2000, batch_size=10)

#Save model
model.save("model.h5")

# Plots 'history'
history_dict=history.history
loss_values = history_dict['loss']
val_loss_values=history_dict['val_loss']
plt.plot(loss_values,'bo',label='training loss')
plt.plot(val_loss_values,'r',label='training loss val')
plt.legend()

# Runs model with its current weights on the training and testing data
y_train_pred = model.predict(X_train_scaled)
y_test_pred = model.predict(X_test_scaled)

# Calculates and prints r2 score of training and testing data
print("The R2 score on the Train set is:\t{:0.3f}".format(r2_score(y_train, y_train_pred)))
print("The R2 score on the Test set is:\t{:0.3f}".format(r2_score(y_test, y_test_pred)))
```
